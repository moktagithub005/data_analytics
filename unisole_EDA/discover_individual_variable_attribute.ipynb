Great — now we move to the second subtopic in the Main Goals of EDA:

 Topic 2: Discover Individual Variable Attributes
This topic focuses on understanding each variable independently, before studying relationships or building models.

This step is called:

Univariate Analysis

Feature-level Understanding

It answers one simple question:

“What does each column look like on its own?”

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("eda_hypothetical_dataset.csv")
```

```python
df.dtypes
```

```python
df.describe()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x=df['Age'])
plt.title("Age Outlier Detection")
plt.show()
```

```python
sns.histplot(df['Units_Sold'])
plt.title("Units Sold Distribution")
plt.show()
```

```python
sns.histplot(df['Income'], kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
df['City'].value_counts()
```

```python
sns.countplot(data=df, y='City')
plt.title("City Frequency")
plt.show()
```

```python
df['Gender'].value_counts()
```

```python
df['Product'].value_counts()
```

```python
df['Promo_Code'].value_counts(dropna=False)
```

```python
df['Fraud_Flag'].value_counts()
```

```python
df['Churned'].value_counts(normalize=True)
```

```python
df['Signup_Date_parsed'] = pd.to_datetime(df['Signup_Date'], errors='coerce')
```

```python
df['Signup_Date_parsed'].describe()
```

```python
df['Signup_Date_parsed'].dt.year.value_counts().sort_index().plot(kind='bar')
```
