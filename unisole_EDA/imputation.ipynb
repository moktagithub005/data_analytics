```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("eda_hypothetical_dataset.csv")
```

```python
df.isnull().sum()
```

```python
mean_age=df['Age'].mean()
```

```python
df['Age_mean_imputed']=df['Age'].fillna(mean_age)
```

```python
## median imputation
median_income=df['Income'].median()
```

```python
df['income_median_imputed']=df['Income'].fillna(median_income)
```

```python
## mode imputation
mode_city=df['City'].mode()
df['city_mode_imputed']=df['City'].fillna(mode_city)
```

```python
##  pip install scikitlearn 
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
```

```python
## select numeric columns for the knn
knn_features=df[['Age','Income','Units_Sold','Total_Amount']]
scaler=StandardScaler()
scaled_feature=scaler.fit_transform(knn_features)
## apply knn imputer
imputer=KNNImputer(n_neighbors=5,weights='uniform')
imputed_scaled=imputer.fit_transform(scaled_feature)
```

```python
imputed_original=scaler.inverse_transform(imputed_scaled)


df_knn_imputed=df.copy()
imputed_original=df_knn_imputed[['Age','Income','Units_Sold','Total_Amount']]
```
