```python
import pandas as pd
df=pd.read_csv("eda_hypothetical_dataset.csv")
```

```python
df['Last_Purchase_Date']=pd.to_datetime(df['Last_Purchase_Date'])
```

```python
df.info()
```

```python
## string to numeric conversion
df["price"]=(
    df["Price"]
    .astype(str)
    .str.replace("Rs","",regex=False)
    .str.strip()
)

df["Price"]=pd.to_numeric(df["Price"],errors="coerce")
```

```python
df.info()
```

```python
df.head()
```

```python
df["Full_Location"]= df["City"]+"_India"
```

```python
df.head()
```

```python
## split columns
df[["Cleaned_city","country"]]=df["Full_Location"].str.split("_",expand=True)
```

```python
##combine columns
df["customer_profile"]=df["Gender"].astype(str)+"_"+df["City"]
```
