```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("eda_hypothetical_dataset.csv")
```

```python
df=df.dropna(subset=["Customer_ID","Signup_Date"])
```

```python
df.head()
```

```python
df.info()
```

```python
median_age=df['Age'].median()
```

```python
print(median_age)
```

```python
df["Age"]=df["Age"].fillna(median_age)
```

```python
df.info()
```

```python
df["Income"]=df["Income"].fillna(df['Income'].median())
```

```python
df.info()
```

```python
df['Gender']=df['Gender'].fillna('Unkown')
```

```python
df.info()
```

```python
threshold=0.4
df=df.loc[:,df.isnull().mean() < threshold]
```

```python
## identify missing value
df[df['Age']>100]
```

```python
df.loc[df['Age'] > 100, 'Age']=df['Age'].median()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(subset=['Customer_ID'],keep="first")
```

```python
df['City'].head()
```

```python
df.describe()
```

```python
df['City']=df['City'].astype(str).str.strip().str.title()
```

```python
df['City'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
sns.boxenplot(x=df['Age'])
plt.title("Age outlier Detection")
plt.xlabel("age")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x="City",y="Total_Amount")
plt.title(" Revenue distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,4))
sns.countplot(dsta=df,x="product")
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="Blues")
plt.title("Correlation Heatmap")
plt.show()
```
