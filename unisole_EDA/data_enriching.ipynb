```python
## data enrichin
import pandas as pd
df=pd.read_csv("eda_hypothetical_dataset.csv")
```

```python
df["Age_Group"]=pd.cut(
    df["Age"],
    bins=[0,18,30,45,60,100],
    labels=["Teen","Young_Adult","Adult","senior","Elder"]
)
```

```python
df["Expected_Revenue"]=df['Units_Sold']*df['Price']
```

```python
df["Revenue_difference"]=df['Total_Amount']-df['Expected_Revenue']
```
