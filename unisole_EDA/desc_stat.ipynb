```python
import pandas as pd
df=pd.read_csv("eda_hypothetical_dataset.csv")
```

```python
df.head()
```

```python
mean=df['Income'].mean()
mean
```

```python
median=df['Income'].median()
median
```

```python
mode=df['Income'].mode()
mode
```

```python
varience=df['Income'].var()
varience
```

```python
std_dev=df['Income'].std()
std_dev
```

```python
range_val=df['Income'].max()-df['Income'].min()
range_val
```

```python
iqr=df['Income'].quantile(0.75)-df['Income'].quantile(0.25)
iqr
```

```python
skewness=df['Income'].skew()
skewness
```

```python
percentile=df['Income'].quantile([0.1,0.25,0.50,0.75,0.90])
percentile
```

```python
grouped=df.groupby('City').agg({
    'Income':['mean','median','std','min','max'],
    'Age':['mean','count']
})
```
