```python
import requests
url="https://jsonplaceholder.typicode.com/posts"
response=requests.get(url)
print(response.status_code)
```

```python
print(type(response))
```

```python
if response.status_code ==200:
    print("data received")

else:
    print("API request failed")
```

```python
data=response.json()
print(type(data))
print(data[0])
```

```python
import pandas as pd
df=pd.DataFrame(data)
```
