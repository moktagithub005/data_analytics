```python
## creating a  series from dict
sdata= {
    "shimla": 35000,
    "chandigrah": 71000,
    "hamirpur": 16000,
    "pune": 5000
}

obj2=pd.Series(sdata)
print(obj2)
```

```python
## forcing thr custom order 
city =["delhi","hamirpur","pune","shimla"]
obj3=pd.Series(sdata,index=city)
print(obj3)
```

```python
pd.isna(obj3)
```

```python
pd.notna(obj3)
```

```python
obj3.dropna()
```

```python
obj3.fillna(0)
```

```python
## automatic Allignment
A=pd.Series([100,50], index=["mon","tue"])
print("Series:\n", A)
```

```python
B=pd.Series([150,50], index=["mon","wed"])
print("Series:\n", B)
```

```python
print(A+B)
```
